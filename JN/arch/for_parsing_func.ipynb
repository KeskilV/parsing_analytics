```python
#выбор папки
level = 0
def fold(sourcepath):
    global level
    if level == 0 :
        folders = os.listdir()
    else:
        folders = os.listdir(sourcepath)
    print([nf for nf in enumerate(folders)])
    case = input (sourcepath)
    if case.isdigit():
        sourcepath += folders[int(case)]+"/"
        level+=1
        sourcepath = fold(sourcepath)
        return sourcepath
    else: 
        print(level, sourcepath)
        return sourcepath
```
