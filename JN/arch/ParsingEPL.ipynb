```python
# рассширение ссылок
def extuls(urls):
    l = []
    for u in urls:
        if u[-1] == '/':
            l.append(u)
        else:
            for i in range(1, int(u[-1])+1):
                l.append(u[:-1]+str(i))
    return l
```

```python
# рассширение ссылок 2
def extuls(urls):
    '''urls = список первой и последней стр'''
    l = []
    for u in urls:
        if u[-1] == '/':
            l.append(u)
        
        else:
            d,a = tuple(u.split('='))
            for i in range(1, int(a)+1):
                l.append(d+'='+str(i))
    return l
```
